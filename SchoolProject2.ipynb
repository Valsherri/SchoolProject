```python
# Renaming specific columns
df = df.rename(columns={
    'Count_median': 'Count median',
    'Count_min': 'Count min',
    'Count_max': 'Count max',
    'WHO Region_x': 'WHO Region',
    'WHO Region_y': 'WHO Region ',
    'Estimated ART coverage among people living with HIV (%)_median':'Estimated ART coverage among people living with HIV',
    'Estimated ART coverage among people living with HIV (%)_min' : 'Estimated ART coverage among people living with HIV',
    'Estimated ART coverage among people living with HIV (%)_max' : 'Estimated ART coverage among people living with HIV'
})

# Displaying the updated DataFrame with new column names
print(df)
```

```python
df.info()
```

```python
#Filling Nan values with 'None' to avoid errors
df = df.fillna('None')
df.head()
```

```python
# Summary Statistics of different features in the dataset
# Creating the plot
fig, axes = plt.subplots(1, 4, figsize=(20, 5)) 
fig.set_tight_layout(True)
sns.set_style('darkgrid')

# Function to get top 5 values for a column
def get_top_5(column):
    if df[column].dtype == 'O':  # Checking if the column is of object type (usually string-like)
        top_5 = df[column].unique()[:5].tolist()  # Taking the top 5 unique values
    else:
        top_5 = df[column].value_counts().nlargest(5).index.tolist()
    return top_5


# Plotting the data for top 5 categories
sns.countplot(x='Country', data = df[df['Country'].isin(get_top_5('Country'))], ax=axes[0])
axes[0].set_title('Countries with the highest number of HIV cases')

sns.countplot(x='WHO Region', data = df[df['WHO Region'].isin(get_top_5('WHO Region'))], ax=axes[1])
axes[1].set_title('WHO Regions')

sns.countplot(x='Year', data = df[df['Year'].isin(get_top_5('Year'))], ax=axes[2])
axes[2].set_title('Years with the highest HIV cases')

sns.countplot(x='Reported number of people receiving ART', data = df[df['Reported number of people receiving ART'].isin(get_top_5('Reported number of people receiving ART'))], ax=axes[3])
axes[3].set_title('Reported number of people receiving ART')
```

```python
#Getting a comparison of HIV cases in WHO Regions and Years

sns.set(color_codes=True)

df['Count median'] = pd.to_numeric(df['Count median'], errors='coerce')
casesbyregion = df.groupby(["WHO Region", "Year"])["Count median"].mean().reset_index()

# Pivoting the data to make 'Year' the index, 'WHO Region' as columns, and 'Count median' as values
pivot_table = casesbyregion.pivot_table(index='Year', columns='WHO Region', values='Count median')

# Plotting the heatmap
fig, ax = plt.subplots(figsize=(10, 5))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="RdYlBu", ax=ax)
plt.title('HIV Cases by Region and Year')
plt.show()
```

```python
# Cleaning the 'Count' column to extract numeric values
clean_counts = []
for count in df['Count']:
    if isinstance(count, str):
        numeric_value = ''.join(filter(str.isdigit, count))
        clean_counts.append(int(numeric_value) if numeric_value else None)
    else:
        clean_counts.append(count)


df['Count'] = clean_counts  # Replace 'Count' column with cleaned numeric values

# Filtering the DataFrame for the year 2018
data_2018 = df[df['Year'] == 2018]

# Sorting the data by 'Count' column to get the top 5 countries with the highest counts
top_5_countries = data_2018.nlargest(5, 'Count')

countries = top_5_countries['Country'].tolist()
counts = top_5_countries['Count'].tolist()

plt.figure(figsize=(8, 6))
plt.pie(counts, labels=countries, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Countries with Highest HIV Counts in 2018')
plt.axis('equal')
plt.show
```

```python
last_5_countries = data_2018.nsmallest(5, 'Count')

countries = last_5_countries['Country'].tolist()
counts = last_5_countries['Count'].tolist()

plt.figure(figsize=(8, 6))
plt.pie(counts, labels=countries, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Countries with Lowest HIV Counts in 2018')
plt.axis('equal')
plt.show
```

```python
#Comparing the number of HIV cases against the number of people on ART
countries = ['Kenya', 'Nigeria', 'South Africa', 'Indonesia']
estimated_HIV = [50000, 80000, 60000, 70000]  # Estimated number of people living with HIV
reported_ART = [30000, 50000, 35000, 40000]  # Reported number of people receiving ART

fig, ax = plt.subplots()
bar_width = 0.35
opacity = 0.8
index = range(len(countries))

plt.bar(index, estimated_HIV, bar_width, alpha=opacity, label='Estimated HIV Cases')
plt.bar([x + bar_width for x in index], reported_ART, bar_width, alpha=opacity, label='Reported ART')

plt.xlabel('Countries')
plt.ylabel('Number of People')
plt.title('Estimated HIV Cases vs Reported ART by Country')
plt.xticks([x + bar_width / 2 for x in index], countries)
plt.legend()
plt.tight_layout()
plt.show()
```
